```
data = pd.read_csv('winequality-white.csv', delimiter=';')
data
```

```
data = data.values
data
```

```
x = data[:,0:11]

#4,898 observations

x_train = x[0:3000]
x_test = x[3000:]
```

```
y = data[:,11]
print(y)
y_train = y[0:3000]
y_test = y[3000:]

from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```
x_train.shape
```

```
y_test.shape
```

```
from keras import models
from keras import layers
```

```
network = models.Sequential()
network.add(layers.Dense(12, activation='relu', input_dim=11))
network.add(layers.Dense(10, activation='relu'))
network.add(layers.Dense(8, activation='relu'))
network.add(layers.Dense(4, activation='relu'))
network.add(layers.Dense(1, activation='softmax'))
```

```
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```
network.fit(x_train, y_train, epochs=70, batch_size=10)
```

```

network.evaluate(x_test, y_test)
```

```
vectors_answers = network.predict(x_test[210:])
vectors_answers = np.where(vectors_answers > 0.5, 1, 0)
print(vectors_answers)
```
